```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#import tensorflow as tf
#import keras as k
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
# to generate a column from another colummn we use "apply" method
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], marker='+', color = 'green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], marker='*', color = 'red')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'], marker='.', color = 'black')
```

```python
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], marker='+', color = 'green')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], marker='*', color = 'red')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'], marker='.', color = 'black')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC()

#model = SVC(C=10) or model = SVC(gamma=1) or model = SVC(kernel='rbf') change the value of C or gamma to tune the model
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
